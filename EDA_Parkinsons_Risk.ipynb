```python
# Checking for dupes
print("Number of Dupes:", df.duplicated().sum())

# Removed dupes
df = df.drop_duplicates()


```

```python
#Had a problem with name so I only selected numbers for now
numeric_df = df.select_dtypes(include=[np.number])

# Calculating numeric columns only
corr = numeric_df.corr()

# Correlation Matrix
print(corr)

# Heatmap
plt.figure(figsize=(16, 14)) 
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler 
from sklearn.svm import SVC 
df = df.dropna()
df.isna().any().to_frame(name='Missing Values')
```

```python
# removed any names 
x = df.drop(['name', 'status'], axis=1)

# Define target variable
y = df['status']

# Split the dataset into training and testing sets (80% train, 20% test)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Standardize the features
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# Import model and accuracy evaluation tools
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Initialize SVM with linear kernel
model = SVC(kernel='linear')

# Train the model
model.fit(x_train_scaled, y_train)

# Predict on test set
y_pred = model.predict(x_test_scaled)

# Evaluate model accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Heatmap visual
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Healthy", "Parkinson's"], yticklabels=["Healthy", "Parkinson's"])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.savefig("confusion_matrix.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 30 Parkinson's cases were correctly identified, but 2 were missed (false negatives).
# 4 healthy cases were correctly labeled, but 3 were misclassified as Parkinson's (false positives).
# Model performs well overall, but slightly over-predicts Parkinsons in healthy individuals.
```
